### 2. Handling Outliers

````python
# Name the Countinuous Variables
continuous_value_column = ['Sales', 'Customers', 'CompetitionDistance', 'CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear', 'DayOfWeek',]
````

````python
# Check the outlier with boxplot
plt.figure(figsize = (30,15))
for n, column in enumerate(continuous_value_column):
  plt.subplot(5,4,n+1)
  sns.boxplot(x =column, data = df).set_title(f'{column.title()}', fontsize = 18)
  plt.tight_layout()
````

````python
# IQR Technique to treat outliers
def iqr2_treat_outliers(data):
  # now calculating first and last quartile
  q1,q3 = np.percentile(data, [25,75])
  # now calculating interquartile range
  IQR = q3-q1
  # now the bounds
  Lower_bound = q1-(1.5*IQR)
  Upper_bound = q3+(1.5*IQR)
  # now the outliers
  Outliers = [x for x in data if x<Lower_bound or x>Upper_bound]
  # replacement with the nearest quartile
  data_treatment = [q1 if x<Lower_bound
                    else q3 if x>Upper_bound
                    else x for x in data]
  int_treateddata = [int (absolute) for absolute in data_treatment]

  return int_treateddata
````

````python
for column in continuous_value_column:
  df[column] = iqr2_treat_outliers(df[column])
````

````python
# now rechecking outliers
# plotting box plot
plt.figure(figsize = (30,15))
for n, column in enumerate(continuous_value_column):
  plt.subplot(5,4,n+1)
  sns.boxplot(x =column, data = df).set_title(f'{column.title()}', fontsize = 18)
  plt.tight_layout()
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

We used IQR Techniques to treat the outliers Because IQR is used to measure variability by dividing a data set into quartiles. The data is sorted in ascending order and split into 4 equal parts.

Text normalization includes techniques like stemming and lemmatization. Stemming reduces words to their base or root form (e.g., "running" to "run"), while lemmatization reduces them to their dictionary form (e.g., "better" to "good"). This aids in achieving consistent representations.

##### Which text normalization technique have you used and why?

Answer Here.

I have knowledge of various text normalization techniques, such as stemming and lemmatization. However, I do not perform text normalization by default. The choice of which technique to use depends on the specific needs of a task.

- **Stemming** involves reducing words to their base or root form. For example, "running" becomes "run". It's computationally less intensive and quicker, but it can sometimes result in non-real words.

- **Lemmatization** reduces words to their dictionary form. For instance, "better" becomes "good". It's more accurate and yields real words, but it can be slower than stemming.

The choice between stemming and lemmatization depends on the specific NLP task. For tasks where preserving the validity and meaningfulness of words is crucial, lemmatization is preferred. For tasks where speed and simplicity are prioritized, stemming may be chosen.

In practical applications, the choice of text vectorization technique depends on the specific NLP task and the characteristics of the dataset. Here are some common scenarios where different techniques might be preferred:

1. **TF-IDF (Term Frequency-Inverse Document Frequency)**:
   - Used when the goal is to weigh the importance of words in a document relative to a larger corpus. It's particularly useful in tasks like information retrieval or document classification.

2. **Word Embeddings (e.g., Word2Vec, GloVe)**:

3. **Bag-of-Words (BoW)**:

4. **N-grams**:

5. **Doc2Vec**:

The choice of text vectorization technique depends on the specific nature of the NLP task, the characteristics of the text data, and the desired outcome. ##### Do you think the dataset is imbalanced? Explain Why.

Answer Here.

A dataset is considered imbalanced when the distribution of classes or labels is significantly skewed, meaning that one class is represented much more than others. This can impact the performance of machine learning models, particularly for tasks where accurate prediction of minority classes is crucial.



**Signs of an Imbalanced Dataset:**

1. **Class Distribution**: One class significantly outnumbers others, e.g., 90% vs. 10% in binary classification.

2. **Visualizations**: Bar or pie charts visually highlight class proportions.

3. **Evaluation Metrics**: If model performance is much better on majority class, it indicates an imbalance.

4. **Domain Knowledge**: Consider context; some imbalances may be expected and not problematic.

5. **Impact on Model**: Imbalances can lead to biased predictions, favoring the majority class.

6. **Resampling Techniques**: Oversampling or undersampling can be used to balance the dataset when needed.

##### What technique did you use to handle the imbalance dataset and why? (If needed to be balanced)

Answer Here.


1. **Resampling**:
   - **Oversampling**: Increasing the representation of the minority class by duplicating or generating synthetic samples. This is helpful when the dataset is too small to capture the characteristics of the minority class.
   - **Undersampling**: Reducing the representation of the majority class by randomly removing samples. This can help reduce the dominance of the majority class.
   - **SMOTE (Synthetic Minority Over-sampling Technique)**: Creates synthetic samples for the minority class, based on the characteristics of existing samples. It's effective in balancing the dataset without introducing duplicate information.

2. **Using Different Algorithms**:
   - Some algorithms are inherently better at handling imbalanced datasets. For example, ensemble methods like Random Forest or Gradient Boosting tend to perform well with imbalanced data.

3. **Cost-sensitive Learning**:
   - Adjusting the misclassification costs during model training to penalize errors on the minority class more heavily. This encourages the model to pay more attention to the minority class.

4. **Ensemble Techniques**:
   - Using techniques like Bagging or Boosting can help improve the performance on the minority class by combining multiple models.

5. **Anomaly Detection**:
   - Treating the minority class as an anomaly detection problem, where the focus is on identifying the rare events.

The choice of technique depends on the specific dataset, the imbalance severity, and the goals of the analysis. It's important to evaluate the performance of the chosen method and consider the trade-offs it introduces. Additionally, it's advisable to experiment with different techniques to find the most effective approach for the specific task at hand.

````python
# Visualizing evaluation Metric Score chart
from sklearn import metrics
def regression_metric (true, predicted):
  Mae = metrics.mean_absolute_error(true, predicted) # Mean absolute error
  Mse = metrics.mean_squared_error(true, predicted) # Mean Squared error
  Rmse = np.sqrt(metrics.mean_squared_error(true, predicted)) # Root mean squre error
  r2_square = metrics.r2_score(true, predicted) # R2 squre
  Mape = mean_absolute_percentage_error(true,predicted)*100 # Mean absolute error
  Accuracy = 100 - Mape
  print('MAE:', Mae)
  print('MSE:', Mse)
  print('RMSE:', Rmse)
  print('MAPE',Mape)
  print('R2 Square', r2_square)
  print('Accuracy ', Accuracy ,'%')
````
